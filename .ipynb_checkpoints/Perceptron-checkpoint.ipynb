### Structure of the Perceptron

The perceptron consists of one or more input units, a summation function, and an activation function. It takes input, weights them, and produces an output based on the activation function.

![perceptron](https://miro.medium.com/v2/resize:fit:940/1*6HtyqOTYyYUzUeERVXL8Ag.png)

---

<div style="text-align: center; margin-top: 40px;">
    <h2 style="font-size: 24px; color: #007bff;">Steps:</h2>
    <ol style="font-size: 18px; text-align: left; margin: auto; width: 60%;">
        <li><b>Initialize Weights:</b> Start with random weights for each input.</li>
        <li><b>Calculate Output:</b> Multiply inputs by weights, sum them up, and pass through an activation function.</li>
        <li><b>Adjust Weights:</b> Update weights based on the error between predicted and actual output.</li>
        <li><b>Repeat:</b> Iterate through steps 2-3 until convergence or a specified number of iterations.</li>
    </ol>
</div>

# Perceptron Trick

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1,random_state=41, hypercube=False, class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr =0.1

    for i in range(1000):
        j = np.random.randint(100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3,2)
```

# Hinge-Loss Perceptron

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    
    w1=w2=b=1
    lr = 0.1
    
    for j in range(1000):
    
        for i in range(X.shape[0]):

            # check condition
            z = w1*X[i][0] + w2*X[i][1] + b

            if z*y[i] < 0:
                w1 = w1 + lr*y[i]*X[i][0]
                w2 = w2 + lr*y[i]*X[i][1]
                b = b + lr*y[i]
            
    return w1,w2,b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
b
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
